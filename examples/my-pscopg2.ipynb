```python
from asyncio import exceptions
import psycopg2

search_han_ji = '雅'
# search_han_ji = '埏'

conn = psycopg2.connect(database="alanjui", user="alanjui", host="127.0.0.1", port="5432")

cur = conn.cursor()

query_str = f"SELECT id, han_ji, chu_im, freq, siann, un, tiau FROM han_ji WHERE han_ji='{search_han_ji}'"
cur.execute(query_str)
rows = cur.fetchall()
for row in rows:
    print(f'ID = {row[0]}')
    print(f'Han_ji = {row[1]}')
    print(f'Chu_im = {row[2]}')
    print(f'Freq = {row[3]}')
    print(f'Siann = {row[4]}')
    print(f'Un = {row[5]}')
    print(f'Tiau = {row[6]}')
    print(f'===========================')

try:
    chu_im = ""
    chu_im = rows[0][2]
except Exception as err:
    print(f"Can not find 【{search_han_ji}】in Han-Ji-Khoo: {err}")

print(f'chu_im = {chu_im}')

print("Operation done sucessfully")
conn.close()
```
